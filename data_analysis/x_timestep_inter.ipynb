```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import seaborn as sns
import numpy as np

pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")

timestep_path = "../organised_output/x_timestep_inter/summary_output.csv"
```

```python
def change_name(name):
    if "++" in name:
        return "greedy insert ++"
    elif "borda" in name:
        return "iterative voting (borda)"
    elif "popularity" in name:
        return "iterative voting (popularity)"
    elif "tsp" in name:
        return "tsp"
    else:
        return "greedy insert"

timestep_df = pd.read_csv(timestep_path, header=0)
timestep_df['algorithm'] = timestep_df['algorithm'].transform(lambda x: change_name(x))
timestep_df.columns = timestep_df.columns.str.replace('_',' ')
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=timestep_df, ax=ax, x='time step', y='avg utilitarian', hue="algorithm", palette=['blue', 'orange', 'green', 'purple'])
ax.set(xlabel="time step (minutes)")
ax.set(ylabel="average sum of utilities")
plt.savefig("../../papers/figures/x_timestep_inter_cluster_y_utilitarian.png")
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=timestep_df, ax=ax, x='time step', y='avg gini index', hue="algorithm", palette=['blue', 'orange', 'green', 'purple'])
ax.set(xlabel="time step (minutes)")
ax.set(ylabel="average gini index")
plt.savefig("../../papers/figures/x_timestep_inter_cluster_y_gini.png")
```
