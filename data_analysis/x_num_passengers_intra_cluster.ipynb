```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import seaborn as sns
import numpy as np

pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")

path = "../organised_output/x_num_passengers_intra_cluster/summary_output.csv"
```

```python
def change_name(name):
    if "++" in name:
        return "greedy insert ++"
    elif "iterative" in name:
        if "borda" in name:
            return "iterative voting (borda)"
        elif "popularity" in name:
            return "iterative voting (popularity)"
    elif "tsp" in name:
        return "tsp"
    else:
        return "greedy insert"

df = pd.read_csv(path, header=0)
df['algorithm'] = df['algorithm'].transform(lambda x: change_name(x))
df.columns = df.columns.str.replace('_',' ')
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=df, ax=ax, x='num passengers', y='avg utilitarian', hue="algorithm", palette=['blue', 'orange', 'green', 'purple'])
ax.set(xlabel="Number of Riders")
ax.set(ylabel="average sum of utilities")
plt.savefig("../../papers/figures/x_numpassengers_intra_cluster_y_utilitarian.png")
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=df, ax=ax, x='num passengers', y='avg gini index', hue="algorithm", palette=['blue', 'orange', 'green', 'purple'])
ax.set(xlabel="Number of Riders")
ax.set(ylabel="average gini index")
plt.savefig("../../papers/figures/x_numpassengers_intra_cluster_y_gini.png")
```
