```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import seaborn as sns
import numpy as np

pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")

gini_path = "../organised_output/x_beta_y_gini/summary_output.csv"
utilitarian_path = "../organised_output/x_beta_y_utilitarian/summary_output.csv"
```

```python
gini_df = pd.read_csv(gini_path, header=0)
gini_df['beta_mean'] = gini_df['alpha'] / (gini_df['alpha'] + gini_df['beta'])
gini_df['algorithm'] = gini_df['algorithm'].transform(lambda x: "greedy insert ++" if "++" in x else "greedy insert")

utilitarian_df = pd.read_csv(utilitarian_path, header=0)
utilitarian_df['beta_mean'] = utilitarian_df['alpha'] / (utilitarian_df['alpha'] + utilitarian_df['beta'])
utilitarian_df['algorithm'] = utilitarian_df['algorithm'].transform(lambda x: "greedy insert ++" if "++" in x else "greedy insert")
```

```python
sns.lineplot(data=gini_df, x='beta_mean', y='avg_gini_index', hue="algorithm", palette=['blue', 'orange'])
plt.savefig("../../papers/figures/x_beta_y_gini.png")
```

```python
sns.lineplot(data=utilitarian_df, x='beta_mean', y='avg_utilitarian', hue="algorithm", palette=['orange', 'blue'])
plt.savefig("../../papers/figures/x_beta_y_utilitarian.png")
```
